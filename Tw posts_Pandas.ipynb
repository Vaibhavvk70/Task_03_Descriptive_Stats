```python
import pandas as pd

#STEP 1: Load Dataset 
df = pd.read_csv("tw_posts_cleaned.csv")
print("Dataset loaded with shape:", df.shape)

#STEP 2: Descriptive Stats for Numeric Columns
numeric_stats = df.describe()
display(numeric_stats)

#STEP 3: Unique Value Counts for All Columns
print("\n=== Unique Value Counts ===")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")

#STEP 4: Most Frequent Values for Non-Numeric Columns
print("\n=== Most Frequent Non-Numeric Values ===")
non_numeric_cols = df.select_dtypes(include='object').columns
for col in non_numeric_cols:
    most_common = df[col].value_counts().head(1)
    print(f"{col}: {most_common.index[0]} ({most_common.values[0]} times)")

#STEP 5: Grouped Stats by 'source'
grouped_by_source = df.groupby("source")[df.select_dtypes(include='number').columns].agg(["count", "mean", "min", "max", "std"])
display(grouped_by_source)

#STEP 6: Grouped Stats by ('source', 'id')
grouped_by_source_id = df.groupby(["source", "id"])[df.select_dtypes(include='number').columns].agg(["count", "mean", "min", "max", "std"])
display(grouped_by_source_id)


```
