```python
import pandas as pd
import json

# STEP 1: Load dataset
df = pd.read_csv("2024_fb_ads_president_scored_anon.csv", low_memory=False)
print("Initial shape:", df.shape)

# STEP 2: Drop rows missing key identifiers
df = df.dropna(subset=["page_id", "ad_id"])

# STEP 3: Convert datetime column
df["ad_creation_time"] = pd.to_datetime(df["ad_creation_time"], errors="coerce")

# STEP 4: Convert numeric columns
numeric_cols = ["estimated_spend", "estimated_impressions", "estimated_audience_size"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# STEP 5: Convert JSON-like string columns
json_cols = ["demographic_distribution", "delivery_by_region"]
for col in json_cols:
    if col in df.columns:
        df[col] = df[col].apply(lambda x: json.loads(x) if pd.notnull(x) and x.startswith("[") else [])

# STEP 6: Normalize string fields
for col in ["currency", "bylines"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.lower()

# STEP 7: Ensure boolean flags are numeric (0 or 1)
flag_cols = [col for col in df.columns if col.endswith("_illuminating")]
for col in flag_cols:
    df[col] = df[col].fillna(0).astype(int)

# Save cleaned main dataset
df.to_csv("fb_ads_cleaned.csv", index=False)
print("Cleaned dataset saved as fb_ads_cleaned.csv")
```

```python
import csv
import math
from collections import defaultdict, Counter

# === CONFIGURATION ===
filename = "fb_ads_cleaned2.csv"

# === UTILITIES ===

def is_number(value):
    try:
        float(value)
        return True
    except:
        return False

def compute_stats(values):
    """Compute count, mean, min, max, stddev for numeric list"""
    count = len(values)
    if count == 0:
        return {"count": 0, "mean": None, "min": None, "max": None, "stddev": None}
    mean = sum(values) / count
    variance = sum((x - mean) ** 2 for x in values) / count
    stddev = math.sqrt(variance)
    return {
        "count": count,
        "mean": mean,
        "min": min(values),
        "max": max(values),
        "stddev": stddev
    }

def summarize_column(values):
    if all(is_number(v) for v in values):
        numeric_vals = [float(v) for v in values]
        return compute_stats(numeric_vals)
    else:
        counter = Counter(values)
        most_common = counter.most_common(1)[0] if counter else (None, 0)
        return {
            "count": len(values),
            "unique": len(counter),
            "most_common": most_common[0],
            "most_common_count": most_common[1]
        }

def aggregate_by(rows, group_keys):
    grouped = defaultdict(list)
    for row in rows:
        key = tuple(row[k] for k in group_keys)
        grouped[key].append(row)
    return grouped

# === LOAD DATASET ===
with open(filename, newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    rows = list(reader)

columns = rows[0].keys()

# === OVERALL SUMMARY ===
print("\n=== OVERALL DATASET SUMMARY ===")
for col in columns:
    values = [row[col] for row in rows if row[col] not in ("", None)]
    summary = summarize_column(values)
    print(f"{col}: {summary}")

# === GROUPED BY page_id ===
print("\n=== SUMMARY GROUPED BY page_id ===")
grouped_by_page = aggregate_by(rows, ["page_id"])

for key, group in list(grouped_by_page.items())[:3]:  # show only first 3 groups
    print(f"\nGroup page_id={key}")
    for col in columns:
        values = [row[col] for row in group if row[col] not in ("", None)]
        summary = summarize_column(values)
        print(f"  {col}: {summary}")

# === GROUPED BY page_id and ad_id ===
print("\n=== SUMMARY GROUPED BY (page_id, ad_id) ===")
grouped_by_page_ad = aggregate_by(rows, ["page_id", "ad_id"])

for key, group in list(grouped_by_page_ad.items())[:3]:  # show only first 3 groups
    print(f"\nGroup page_id, ad_id={key}")
    for col in columns:
        values = [row[col] for row in group if row[col] not in ("", None)]
        summary = summarize_column(values)
        print(f"  {col}: {summary}")
```

```python
import polars as pl

# === Load the cleaned dataset ===
df = pl.read_csv("fb_ads_cleaned2.csv")
print("Data loaded. Shape:", df.shape)

# === Separate numeric and non-numeric columns ===
numeric_cols = [col for col in df.columns if pl.datatypes.is_numeric(df[col].dtype)]
non_numeric_cols = [col for col in df.columns if col not in numeric_cols]

# === OVERALL NUMERIC SUMMARY ===
print("\n=== OVERALL SUMMARY (numeric columns) ===")
print(df.select(numeric_cols).describe())

# === OVERALL NON-NUMERIC SUMMARY ===
print("\n=== OVERALL SUMMARY (non-numeric columns) ===")
for col in non_numeric_cols:
    unique_count = df[col].n_unique()
    most_common = df.groupby(col).count().sort("count", descending=True).first()
    print(f"\nColumn: {col}")
    print(f"Unique values: {unique_count}")
    print(f"Most common value: {most_common[col].item()} (count: {most_common['count'].item()})")

# === GROUPED BY page_id ===
print("\n=== SUMMARY GROUPED BY page_id ===")
grouped_page = df.groupby("page_id").agg([
    *[pl.col(c).mean().alias(f"{c}_mean") for c in numeric_cols],
    *[pl.col(c).min().alias(f"{c}_min") for c in numeric_cols],
    *[pl.col(c).max().alias(f"{c}_max") for c in numeric_cols],
    *[pl.col(c).std().alias(f"{c}_std") for c in numeric_cols]
])
print(grouped_page.head(3))

# === GROUPED BY (page_id, ad_id) ===
print("\n=== SUMMARY GROUPED BY (page_id, ad_id) ===")
grouped_page_ad = df.groupby(["page_id", "ad_id"]).agg([
    *[pl.col(c).mean().alias(f"{c}_mean") for c in numeric_cols],
    *[pl.col(c).min().alias(f"{c}_min") for c in numeric_cols],
    *[pl.col(c).max().alias(f"{c}_max") for c in numeric_cols],
    *[pl.col(c).std().alias(f"{c}_std") for c in numeric_cols]
])
print(grouped_page_ad.head(3))
```
