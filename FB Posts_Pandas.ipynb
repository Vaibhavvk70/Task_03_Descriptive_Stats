```python
import pandas as pd

#STEP 1: Load the dataset
df = pd.read_csv("fb_posts_cleaned.csv")
print("Dataset loaded. Shape:", df.shape)

#STEP 2: Descriptive statistics for numeric columns
print("\n=== Overall Descriptive Statistics (numeric) ===")
print(df.describe())

#STEP 3: Unique counts for all columns
print("\n=== Number of Unique Values per Column ===")
print(df.nunique())

#STEP 4: Most frequent values for non-numeric columns
print("\n=== Most Frequent Values (Top 1) for Non-Numeric Columns ===")
for col in df.select_dtypes(include=["object"]).columns:
    most_common = df[col].value_counts().head(1)
    print(f"{col}: {most_common.index[0]} (count: {most_common.iloc[0]})")

#STEP 5: Grouped Stats by Facebook_Id (Only numeric columns)
numeric_cols = df.select_dtypes(include="number").columns

print("\n=== Descriptive Statistics Grouped by Facebook_Id ===")
grouped_page = df.groupby("Facebook_Id")[numeric_cols].agg(["count", "mean", "min", "max", "std"])
print(grouped_page.head(3))

#STEP 6: Grouped by Facebook_Id and post_id (Only numeric columns)
print("\n=== Descriptive Statistics Grouped by Facebook_Id and post_id ===")
grouped_page_post = df.groupby(["Facebook_Id", "post_id"])[numeric_cols].agg(["count", "mean", "min", "max", "std"])
print(grouped_page_post.head(3))
```
